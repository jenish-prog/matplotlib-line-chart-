```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Basic graph

```python
x = [0,1,2,3]
y = [0,4,5,6]

#resize your graph
plt.figure(figsize=(5,3),dpi=100)

#plt.plot(x,y, label='2x', color = 'brown', linewidth = '10', marker = '.', linestyle = '--', markersize = '30', markeredgecolor = 'green')

#use shorthand notation
#fmt = '[color][marker][line]'
plt.plot(x,y, 'b.--', label = '2x')

#line number two
#select the interval we want to plot points at
x2 = np.arange(0,4.5,0.5)

#plot part of the as line
plt.plot(x2[:6],x2[:6]**2, 'r', label='x**2')

#plot reminder of the graph
plt.plot(x2[5:],x2[5:]**2,'r--')

#adding title to graph
plt.title('Line Plot' , fontdict = {'fontname':'Comic Sans MS' , 'fontsize' : 20})

##x and y labels
plt.xlabel('x - axis')
plt.ylabel('y - axis')

#x and y tick marks(to avoid the boolean scale numbers)
plt.xticks([0,1,2,3])
plt.yticks([0,4,5,6])

#save a graph
plt.savefig('linegraph.png',dpi=300)

plt.show()
```

### Bar chart

```python
labels = ['A','B','C']
values = [5,10,15]

bars = plt.bar(labels,values)

patterns = ['/','o','*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

#bars[0].set_hatch('/')
#bars[1].set_hatch('o')
#bars[2].set_hatch('*')

plt.figure(figsize=(5,6))

plt.show()
```
